```python
# Convert 'Memory' to string, remove '.0', 'GB', and convert 'TB' to '000' (GB equivalent)
df['Memory'] = df['Memory'].astype(str).str.replace(r'\.0', '', regex=True)
df['Memory'] = df['Memory'].str.replace('GB', '', regex=False).str.replace('TB', '000', regex=False)

# Split memory into two parts if multiple storage types exist
df[['first', 'second']] = df['Memory'].str.split("+", n=1, expand=True).fillna("0")

# Identify storage types and extract numeric values
storage_types = ['HDD', 'SSD', 'Hybrid', 'Flash Storage']
for storage in storage_types:
    df[storage] = df['first'].str.contains(storage, na=False).astype(int) * df['first'].str.extract(r'(\d+)')[0].fillna(0).astype(int) + \
                  df['second'].str.contains(storage, na=False).astype(int) * df['second'].str.extract(r'(\d+)')[0].fillna(0).astype(int)

# Drop temporary columns
df.drop(columns=['first', 'second', 'Memory'], inplace=True)

# Display final dataset
print(df.head())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.model_selection import train_test_split, RandomizedSearchCV

import warnings
warnings.filterwarnings('ignore', category=UserWarning)
```

```python
 # Linear Regression

step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])
], remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
print('Linear Regression RMSE:', rmse)
```

```python

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=2)

# Preprocessing (OneHotEncoding for categorical columns)
preprocessor = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Define Models
models = {
    "RandomForest": RandomForestRegressor(random_state=42),
    "GradientBoosting": GradientBoostingRegressor(random_state=42)
}

# Hyperparameter Grids
param_grids = {
    "RandomForest": {
        'model__n_estimators': [50, 100, 200],
        'model__max_depth': [10, 20, 30, None],
        'model__min_samples_split': [2, 5, 10],
        'model__min_samples_leaf': [1, 2, 4],
        'model__max_features': ['sqrt', 'log2']
    },
    "GradientBoosting": {
        'model__n_estimators': [100, 200, 300],
        'model__learning_rate': [0.01, 0.05, 0.1],
        'model__max_depth': [3, 5, 7],
        'model__min_samples_split': [2, 5, 10],
        'model__min_samples_leaf': [1, 2, 4],
        'model__subsample': [0.7, 0.8, 1.0]
    }
}

# Train & Tune Both Models
for name, model in models.items():
    print(f"\nTuning {name} Model...")
    
    # Create Pipeline
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    # RandomizedSearchCV
    search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=param_grids[name],
        n_iter=20,
        cv=5,
        scoring='r2',
        verbose=2,
        random_state=42,
        n_jobs=-1
    )
    
    # Fit Model
    search.fit(X_train, y_train)
    
    # Best Parameters
    print(f"\n{name} Best Parameters:", search.best_params_)
    
    # Best Model & Predictions
    best_model = search.best_estimator_
    y_pred = best_model.predict(X_test)
    
    # Evaluation
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    print(f' Best Tuned {name} R: {r2:.4f}')
    print(f' Best Tuned {name} RMSE: {rmse:.4f}')
```

```python

from sklearn.ensemble import GradientBoostingRegressor

# Create final model with best parameters
final_model = GradientBoostingRegressor(
    subsample=0.7,
    n_estimators=100,
    min_samples_split=5,
    min_samples_leaf=4,
    max_depth=7,
    learning_rate=0.05
)


from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ('step1', preprocessor), 
    ('step2', final_model)   
])

# Fit pipeline with training data
pipe.fit(X_train, y_train)

import pickle

# Save the trained model (pipeline)
with open("pipe.pkl", "wb") as f:
    pickle.dump(pipe, f)

print(" pipe.pkl saved successfully!")

# Save the dataframe
with open("df.pkl", "wb") as f:
    pickle.dump(df, f)

print(" df.pkl saved successfully!")

```

```python
# Step 1: Get the trained model and feature names
final_model = pipe.named_steps['step2']  # This is  trained model (RandomForest or GradientBoosting)
feature_names = pipe.named_steps['step1'].get_feature_names_out()  # Names after encoding & transformation

# Step 2: Create DataFrame with importances
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': final_model.feature_importances_
})

def clean_feature_names(feature_name):
    return re.sub(r'(remainder__|col_tnf__)', '', feature_name)

importance_df['Feature'] = importance_df['Feature'].apply(clean_feature_names)

# Step 4: Sort by importance and select top 15
top_features = importance_df.sort_values(by='Importance', ascending=False).head(15)

# Step 5: Plot combined graphs - Top 15, Numerical, Categorical Importance
plt.figure(figsize=(12, 18))

# Top 15 Features
plt.subplot(3, 1, 1)
plt.barh(top_features['Feature'], top_features['Importance'], color='skyblue')
plt.title('Top 15 Feature Importance')
plt.xlabel('Importance')
plt.gca().invert_yaxis()

# Numerical Features Importance
numerical_features = [f for f in importance_df['Feature'] if f in ['Ram (In GBs)', 'SSD', 'Weight (In Kgs)', 'HDD', 'ppi', 'Touchscreen', 'Ips']]
numerical_importance = importance_df[importance_df['Feature'].isin(numerical_features)]

plt.subplot(3, 1, 2)
plt.barh(numerical_importance['Feature'], numerical_importance['Importance'], color='lightgreen')
plt.title('Numerical Feature Importance')
plt.xlabel('Importance')
plt.gca().invert_yaxis()

# Categorical Features Importance (remaining)
categorical_importance = importance_df[~importance_df['Feature'].isin(numerical_features)]

plt.subplot(3, 1, 3)
top_categorical = categorical_importance.sort_values(by='Importance', ascending=False).head(15)
plt.barh(top_categorical['Feature'], top_categorical['Importance'], color='lightcoral')
plt.title('Categorical Feature Importance')
plt.xlabel('Importance')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
